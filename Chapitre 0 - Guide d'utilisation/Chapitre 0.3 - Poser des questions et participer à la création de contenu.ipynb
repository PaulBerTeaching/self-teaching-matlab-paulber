# Chapitre 0.3 - Poser des questions et participer à la création de contenu

La plateforme que vous utilisez à également un côté participatif : vous avez la possibilité de poser des questions aux enseignants qui ont participé à la création du contenu. 

Pour cela, vous devez vous rendre sur la page GitHub du projet : <https://github.com/paulber/MiseANiveauMatlab>. 

## Poser des questions

Pour poser des questions et suggérer des modifications : chaque projet GitHub présente un onglet "issue" qui permet à des utilisateurs extérieurs de rédiger des requêtes ou exposer des bugs. 



Vous pouvez utiliser cet espace pour nous écrire vos suggestions et problèmes (mathématiques, informatiques ou d'utilisation).Si vous souhaitez une réponse plus rapide, vous pouvez également nous contacter par mail. Voici les adresses :

## Participer à la création

Pour participer : vous pouvez *fork* le projet et rajouter les éléments que vous jugez important pour développer les différents cours. Une fois vos modifications faites sur votre machine, vous devez créer un *pull request* pour ajouter vos propositions au répertoire principal. Les fichiers README.md et CONTRIBUTION.md n'étant pas rédigés à l'heure actuel, vous pouvez consulter cette page pour installer les outils nécessaires : <https://datascience-enthusiast.com/Miscellaneous/Jupyter_R_Python_Julia_Octave.html>
