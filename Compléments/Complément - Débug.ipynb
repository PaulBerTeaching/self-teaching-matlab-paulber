# Complément - Débug

L'éditeur Matlab est suffisamment puissant pour vous permettre de corriger votre code au fur et à mesure de son exécution. il faut simplement lui dire où s'arrêter pour que vous puissiez faire des tests. Une fois que vous aurez effectué le bon test et identifié le problème, vous devez vous poser les bonnes questions pour corriger ce qui pose problème. Cette session est faite pour vous donner les outils qui vous débloqueront lorsque vous aurez des erreurs.

## Les breakpoints

Vous avez dû remarquer que lorsque vous faites tourner un script, Matlab bloque le terminal et n'exécutera les commandes que vous saisissez que lorsqu'il aura fini d'interpréter le script. Si vous avez besoin d'exécuter du code intermédiaire pour tester une partie de script ou rentrer dans une fonction lorsque le script l'appelle, vous pouvez utiliser les breakpoint pour mettre en pause l'exécution.

Les breakpoints se placent facilement, dans l'éditeur Matlab vous pouvez apercevoir une petite colonne à côté des numéros de lignes contenant de petits traits. Il vous suffit de cliquer sur un de ces petits traits pour placer un breakpoint (représenté sous forme d'une boule).



Une fois le script lancé, Matlab s'arrétera automatiquement au point que vous avez fixé et le terminal redeviendra disponible pour exécuter du code en ligne par ligne. Vous remarquerez que le chevron s'est changé en *K>>*.

Pour reprendre l'exécution, il vous faudra cliquer sur "Continue"pour aller jusqu'au prochain Breakpoint ou du script.

**NB :** Vous pouvez très bien appuyer sur "pause" pendant l'exécution du script, mais vous risquez de vous arréter n'importe où.

## Petites astuces pour se débloquer

  Généralement, lorsque vous ne trouvez pas une solution au bout de 2 à 5 minutes (ce temps peut varier), vous levez la main et patientez. En fonction du déroulement de la séance, vous pouvez attendre plus d'une dizaine de minutes pour une histoire de variable mal écrite. Afin de faciliter le travail pour tous, nous vous proposons un liste de piste à suivre pour vous débloquer : 
  
  - Lire l'erreur affichée et essayer de la comprendre : si Matlab vous dit qu'il ne connaît pas la variable, c'est parce qu'elle n'est pas déclarée ou alors le nom n'est pas correctement écrit;
  - Utiliser Internet si vous ne comprenez toujours pas l'erreur pour voir quelle autre solution existe;
  - Vos résultats vous semblent incohérents ? Relire l'énoncé, il est peut-être logique d'avoir un mauvais résultat et on attend de vous une réflexion mathématique;
  - Toujours rien et vous avez clairement identifié la fonction qui pose problème : vérifier le type et le contenu des paramètres d'entrée peut résoudre le problème (des NaN dans un vecteur, un paramètre nul alors qu'il ne devrait pas);
  - Si ce n'est pas un problème de type/contenu de paramètre, peut-être que sa valeur entraîne mathématiquement l'échec de l'algorithme. Revoir les conditions de réussite de l'algorithme et vérifier quelles sont respectées;
  
Rappelez-vous toujours : si le problème n'est pas informatique, il est mathématiques. Vous devez donc repartir du papier et du crayon pour voir ce qui ne va pas. Et si le problème n'est pas mathématiques alors il est informatique et vous devez revoir votre systême d'implémentation. Enfin, si ce n'est ni l'un, ni l'autre et que le problème est présent chez plusieurs groupes ( et à condition que le code n'ait pas juste été massivement partagé) : alors le problème doit venir de l'énoncé.
