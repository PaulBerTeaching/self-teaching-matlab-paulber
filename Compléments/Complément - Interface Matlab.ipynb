# Interface Matlab : les astuces

L'inconvénient avec cette méthode d'apprentissage, c'est que vous ne pouvez pas travailler directement avec l'interface graphique de Matlab. C'est pourquoi, nous recommandons à ceux qui seront nouveaux au langage lors du suivi de ces cours, d'essayer le logiciel sur une des machines de l'université avant d'aller à leur premier TP. 

Concernant l'interface graphique de Matlab, il se présente de la manière suivante :



Vous pouvez constater la présence de plusieurs zones : 

 - La partie encadrée en bleu (= l'en tête) : contient un ensemble de vignettes et d'outils dans une barre des tâches avec en dessous le chemin du dossier dans lequel vous vous trouvez;
 - En haut à gauche (" Current folder ") : le contenu du dossier courant ;
 - En bas à gauche (" Details ") : des détails sur le contenu du fichier sélectionné, généralement des lignes de commentaires ;
 - Au centre : un terminal qui est la zone dans laquelle vous saisirez vos commandes. Le chevron >> indique que Matlab attend vos instructions ;
 - En haut à droite (" Workspace ") : les variables d'environnement avec leurs noms et leur valeur ;
 - En bas à droite (" Command history ") : l'historique des commandes saisies.
 
 ## La fenêtre détails
 
 La fenêtre de détail permet d'afficher les commentaires précédés de deux symboles *%*. Cette fonctionnalité est pratique lorsque l'on souhaite savoir rapidement ce que le fichier contient. Pour l'enseignant, cela permet aussi de savoir quel étudiant à créer le contenu de la fonction si vous avez commenté de la manière suggérée.
 
 ## Le terminal
 
 C'est la zone qui s'apparente à celle que vous avez utilisé dans l'ensemble des notebooks à l'exception que les lignes sont exécutées au fur et à mesure. Il s'agit d'une zone de test rapide : vous voulez exécuter une fonction sans avoir à garder une trace dans un script alors utilisez le terminal.
 
 Dans le terminal, vous pouvez naviguer dans les commandes que vous avez déjà saisies en appuyant sur les flèches 'haut' et 'bas' de votre clavier.
 
 **Attention :** Il n'est pas judicieux de faire l'ensemble d'un TP dans un terminal. Rien ne vous assure que votre ordinateur ne plantera pas en cours de route ! Il serait dommage de faire une heure de TP sans sauvegarder au fur et à mesure les lignes de commandes dans un script et de devoir tout réexécuter ligne par ligne pour revenir au point où vous en étiez.
 
 ## Le Workspace
 
 Il s'agit d'une zone qui affiche les noms des variables déjà créées et leur valeur (ou leur type dans bien des cas). Il s'agit d'un aide-mémoire rapide pour savoir où vous en êtes dans vos fonctions, un double-clic sur la variable permet d'afficher son contenu.  Vous pouvez aussi manuellement supprimer, renommer ou modifier des variables.
 
 ## L'historique de commandes
 
 Comme son nom l'indique, il affiche les différentes commandes que vous avez saisies. Vous pouvez cliquer-glisser les commandes dans votre terminal ou dans vos scripts/fonctions.
 
 ## L'en-tête
 
L'en-tête Matlab est subdivisée en petits onglets. Ces onglets sont au nombre de 3 à la base et passent à 6 si vous ouvrez l'éditeur de script/fonction. Si votre éditeur est en dehors de la fenêtre principale (ce qui peut être utile pour programmer), vous revenez au nombre d'onglet de base avec 3 onglets dans la nouvelle fenêtre d'édition. De manière général, voici les rôles des onglets :

- **Home :** onglet principal qui vous permet de créer en quelques clics de nouveaux scripts/fonctions, de nettoyer votre environnement de travail, de modifier votre affichage, ...;
- **Plots :** qui offre une assistance pour afficher une fonction. Elle peut être utile pour démarrer, mais rapidement vous préférerez saisir les commandes manuellement;
- **Apps :** Matlab propose un ensemble d'application pour atteindre un public plus large que les mathématiciens. Les apps disponibles varient de votre installation, mais on peut notamment avoir des outils de compilations, de déploiement,...;
- **Editor :** Contient des outils d'édition de commentaire ou de mise en forme de votre code (*edit*), les options sur les breakpoints (voir le cours complémentaire *Débug*) et les options d'exécutions de script;
- **Publish :** Contient les options pour la création de la documentation
- **View :** Contient les options pour modifier l'affichage

À présent, vous avez fait le tour de l'interface graphique. Une fois habitué, vous remarquerez que cela peut vous faciliter la tâche dans la création de vos programmes.
